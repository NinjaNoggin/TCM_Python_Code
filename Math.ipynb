```python
#!/bin/python3

#MATH
print(50 + 50) #Addition
print(50 - 50) #Subtraction
print(50 * 50) #Multiplication
print(50 / 50) #Division
print(50 + 50 - 50 * 50 / 50) #Order of operations
print(50 ** 2) #Exponentiation
print(50 % 6) #Modulus (remainder)
print(50 / 6) #Division (float)
print(50 // 6) #Floor division (integer)(no remainder)
```
